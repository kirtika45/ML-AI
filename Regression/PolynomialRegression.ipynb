# **Polynomial** **Regression**

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
data_set = pd.read_excel("RBC_in_humans.xlsx")
X = data_set.iloc[:, :-1].values
y = data_set.iloc[:, -1].values
```

### Training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

### Training the Polynomial Regression on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_reg = PolynomialFeatures(degree = 6)
X_poly = polynomial_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

### Visualization using Linear Reg.

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('RBC in Humans')
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

### Visualization using Polynomial Reg.

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly),color='blue')
plt.title('RBC in Humans')
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

### Smooth graph

```python
x2 = np.arange(min(X),max(X),0.1)
x2 = x2.reshape((len(x2)),1)
plt.scatter(X,y,color='red')
plt.plot(x2,lin_reg_2.predict(polynomial_reg.fit_transform(x2)),color='blue')
plt.show()
```

### Prediction

```python
prediction = polynomial_reg.fit_transform([[22]])
lin_reg_2.predict(prediction)
```
